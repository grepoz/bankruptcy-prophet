```python
import numpy as np
import pandas as pd
```

```python
train_ratio = 0.8
validation_ratio = 0.05
test_ratio = 0.15
```

```python
synthetic_financial_positive_data = pd.read_csv('synthetic_financial_positive_data.csv')
synthetic_financial_positive_data.drop(['subset'], axis=1, inplace=True)
synthetic_financial_positive_data.head()
```

```python
synthetic_textual_positive_data = pd.read_csv('synthetic_textual_data.csv')
synthetic_textual_positive_data.head()
```

```python
synthetic_financial_positive_data_grouped = synthetic_financial_positive_data.groupby('cik').agg({'label': 'first'}).reset_index()
fin_data_length = len(synthetic_financial_positive_data_grouped)

train_set = synthetic_financial_positive_data_grouped[:int(train_ratio * fin_data_length)]
validation_set = synthetic_financial_positive_data_grouped[int(train_ratio * fin_data_length):int((train_ratio + validation_ratio) * fin_data_length)]
test_set = synthetic_financial_positive_data_grouped[int((train_ratio + validation_ratio) * fin_data_length):]

fin_data_length, len(train_set), len(validation_set), len(test_set)
```

```python
train_set.head()
```

```python
train_set['subset'] = 'train'
validation_set['subset'] = 'val'
test_set['subset'] = 'test'

train_set.drop(columns=['label'], inplace=True)
validation_set.drop(columns=['label'], inplace=True)
test_set.drop(columns=['label'], inplace=True)

subsets = pd.concat([train_set, validation_set, test_set])
subsets.head()
```

```python
subsets = subsets.sample(frac=1).reset_index(drop=True)
subsets.head()
```

```python
synthetic_financial_positive_data = subsets.merge(synthetic_financial_positive_data, on='cik', how='left')
synthetic_financial_positive_data.head()
```

```python
synthetic_financial_positive_data['subset'].value_counts(normalize= True)
```

```python
synthetic_financial_positive_data.to_csv('synthetic_financial_positive_data_split.csv', index=False)
```

```python
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports = pd.read_csv('bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports.csv')
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports.head()
```

```python
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports['subset'].value_counts(normalize=True)
```

```python
len(bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports)
```

```python
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports = pd.concat([bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports, synthetic_financial_positive_data])
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports.head()
```

```python
len(bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports)
```

```python
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports['subset'].value_counts(normalize=True)
```

```python
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_grouped = bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports.groupby('cik').agg({'label': 'first'}).reset_index()

# shuffle the data
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_grouped_shuffled  = bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_grouped.sample(frac=1)

# ungroupped data
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled = bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_grouped_shuffled .merge(bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports, on='cik', how='left')

bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled.drop(columns=['label_y'], inplace=True)
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled.rename(columns={'label_x': 'label'}, inplace=True)

bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled.head()
```

```python
bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled.to_csv('bankrupt_companies-17_variables-5_years_per_object-balanced-split-matched_with_textual_data-shuffled.csv', index=False)
```

### Textual Data - splitting and matching

```python
fin_data_to_merge = bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled.groupby('cik').agg({'label': 'first', 'subset': 'first'})
print(len(fin_data_to_merge))
fin_data_to_merge.head()
```

```python
synthetic_textual_data = pd.read_csv('synthetic_textual_data.csv')
print(len(synthetic_textual_data))
synthetic_textual_data.head()
```

```python
textual_data_matched_with_fin_data_with_labels_split = pd.read_csv('textual_data_matched_with_fin_data_with_labels_split.csv')
print(len(textual_data_matched_with_fin_data_with_labels_split))
textual_data_matched_with_fin_data_with_labels_split.head()
```

```python
textual_data_matched_with_fin_data_with_labels_split.drop(columns=['label', 'subset'], inplace=True)
```

```python
final_textual_data = pd.concat([textual_data_matched_with_fin_data_with_labels_split, synthetic_textual_data])
print(len(final_textual_data))
final_textual_data.head()
```

```python
final_textual_data_merged = fin_data_to_merge.merge(final_textual_data, on='cik', how='left')
print(len(final_textual_data_merged))
final_textual_data_merged.head()
```

```python
final_textual_data_merged.to_csv('textual_data-matched_with_fin_data-split.csv', index=False)
```

```python
# get every fifth row from 
every_fifth_index = bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled.index % 5 == 0 
every_fifth_data = bankrupt_companies_with_17_variables_5_years_version2_split_matched_with_reports_shuffled[every_fifth_index].reset_index(drop=True)

every_fifth_data.head()
```

```python
every_fifth_data_cik = every_fifth_data['cik']
every_fifth_data_cik.head()
```

```python
# order final_textual_data_merged by every_fifth_data_cik
final_textual_data_merged_copy = final_textual_data_merged.copy()
final_textual_data_merged_copy['cik'] = pd.Categorical(final_textual_data_merged_copy['cik'], categories=every_fifth_data_cik, ordered=True)

# Step 5: Sort the DataFrame based on this categorical column
final_textual_data_merged_copy = final_textual_data_merged_copy.sort_values('cik')

# Reset index if needed
final_textual_data_merged_copy.reset_index(drop=True, inplace=True)

final_textual_data_merged_copy.head()
```

```python
final_textual_data_merged_copy.to_csv('textual_data-matched_and_aligned_with_fin_data-split.csv', index=False)
```
